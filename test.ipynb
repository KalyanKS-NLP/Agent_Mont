```python
import google.generativeai as genai
from agent_mont import AgentMontExtended
import os

# Configure API Key
os.environ["GOOGLE_API_KEY"] = ""
genai.configure(api_key=os.getenv("GOOGLE_API_KEY"))
```

```python
# Initialize Agent Mont Monitoring
mont = AgentMontExtended(model="gemini-1.5-flash")
mont.start()

# Define input prompt
input_text = "Write a short poem about AI and the future."

# Use Google Gemini for text generation
model = genai.GenerativeModel("gemini-1.5-flash")
response = model.generate_content(input_text)
output_text = response.text
```

```python
# Log token usage
mont.set_token_counts(input_text=input_text, output_text=output_text)
mont.end()

# Display output
print("\nGenerated Response:")
print(output_text)

# Visualize Monitoring Metrics
mont.visualize(method='cli')
```

```python

from langchain_google_genai import ChatGoogleGenerativeAI
import time
from langchain.schema import SystemMessage, HumanMessage
from agent_mont import AgentMontExtended

# Initialize Agent Mont for monitoring
mont = AgentMontExtended(model="gemini-1.5-flash")
mont.start()

# Initialize Google Gemini Pro model
llm = ChatGoogleGenerativeAI(model="gemini-1.5-flash")

# Define a simple agent conversation
messages = [
    SystemMessage(content="You are a helpful AI assistant."),
    HumanMessage(content="What are the benefits of AI in supply chain management?")
]

# Start measuring execution time
start_time = time.time()
response = llm(messages)
end_time = time.time()

# Log execution time
mont.set_token_counts(input_text=messages[1].content, output_text=response.content)
mont.end()

# Print response
print("\n### AI Response:")
print(response.content)

# Visualize monitoring results
mont.visualize(method='cli')
```
